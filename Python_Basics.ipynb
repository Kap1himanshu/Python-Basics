# 2. What is an interpreter in Python?

# --------------------------------------------------

# An interpreter is a program that executes Python code

# line by line.

# Unlike a compiler, it does not convert the entire program

# into machine code at once.

# Python uses an interpreter, which makes debugging easier.

# 3. What are pre-defined keywords in Python?

# --------------------------------------------------

# Pre-defined keywords are reserved words in Python

# that have special meaning.

# Examples:

# if, else, for, while, True, False, None, def, return, class

# These cannot be used as variable names.

# --------------------------------------------------

# 4. Can keywords be used as variable names?

# --------------------------------------------------

# No, keywords cannot be used as variable names

# because they are reserved for special purposes.

# Example:

# You cannot write: if = 10  (This will give syntax error)

# 5. What is mutability in Python?

# --------------------------------------------------

# Mutability refers to whether an object can be changed

# after it is created.

# Mutable objects can be modified.

# Immutable objects cannot be modified.

# --------------------------------------------------

# 6. Why are lists mutable, but tuples are immutable?

# --------------------------------------------------

# Lists are mutable because they are designed to allow

# dynamic changes such as adding, removing, or modifying elements.

# Tuples are immutable to ensure data integrity and faster performance.

# Once created, tuple elements cannot be changed.

# --------------------------------------------------

# 7. What is the difference between "==" and "is" operators?

# --------------------------------------------------

# == checks whether values are equal.

# is checks whether two variables refer to the same memory location.

# Example:

# a = [1,2]

# b = [1,2]

# a == b  -> True  (values equal)

# a is b  -> False (different memory location)

# --------------------------------------------------

# 8. What are logical operators in Python?

# --------------------------------------------------

# Logical operators are used to combine conditional statements.

# The three logical operators are:

# and  -> Returns True if both conditions are True

# or   -> Returns True if at least one condition is True

# not  -> Reverses the logical state

# 9. What is type casting in Python?

# --------------------------------------------------

# Type casting is the process of converting one data type

# into another data type.

# Example:

# Converting string to integer, integer to float, etc.

# Python provides built-in functions like int(), float(), str(), bool().

# --------------------------------------------------

# 10. What is the difference between implicit and explicit type casting?

# --------------------------------------------------

# Implicit Type Casting:

# Done automatically by Python.

# Example:

# num = 5 + 2.5   -> Result becomes float automatically

# Explicit Type Casting:

# Done manually by the programmer.

# Example:

# num = int("10")  -> Converts string to integer

# --------------------------------------------------

# 11. What is the purpose of conditional statements in Python?

# --------------------------------------------------

# Conditional statements are used to make decisions in a program.

# They execute specific blocks of code based on conditions.

# Example:

# if, elif, else

# They allow dynamic behavior in programs.

# --------------------------------------------------

# 12. How does the elif statement work?

# --------------------------------------------------

# The elif (else if) statement is used to check multiple conditions.

# If the first condition is False,

# Python checks the elif condition.

# It helps avoid multiple nested if statements.

# Example:

# if x > 0:

# print("Positive")

# elif x == 0:

# print("Zero")

# else:

# print("Negative")

# --------------------------------------------------

# 13. What is the difference between for and while loops?

# --------------------------------------------------

# for loop:

# Used when the number of iterations is known.

# Example:

# for i in range(5):

# while loop:

# Used when the number of iterations is not known.

# Runs as long as a condition is True.

# --------------------------------------------------

# 14. Describe a scenario where a while loop is more suitable than a for loop.

# --------------------------------------------------

# A while loop is more suitable when the number of repetitions

# depends on a condition rather than a fixed range.

# Example:

# Asking user to enter password until correct one is entered.

# The loop continues until the condition becomes False.

```python
# ==================================================
# ==================================================

# --------------------------------------------------
# 6. Function to demonstrate mutable and immutable arguments
# --------------------------------------------------

def modify_values(num, lst):
    num = num + 10          # Immutable (int will not change outside)
    lst.append(4)           # Mutable (list will change outside)
    print("Inside Function - num:", num)
    print("Inside Function - list:", lst)

number = 5
my_list = [1, 2, 3]

modify_values(number, my_list)

print("Outside Function - number:", number)
print("Outside Function - list:", my_list)


# --------------------------------------------------
# 7. Basic arithmetic operations on two numbers
# --------------------------------------------------

a = 10
b = 5

print("\nArithmetic Operations:")
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Modulus:", a % b)


# --------------------------------------------------
# 8. Demonstrate logical operators
# --------------------------------------------------

x = 10
y = 20

print("\nLogical Operators:")
print("x > 5 and y > 15:", x > 5 and y > 15)
print("x > 15 or y > 15:", x > 15 or y > 15)
print("not(x > 5):", not(x > 5))


# --------------------------------------------------
# 9. Convert string input to int, float, and bool
# --------------------------------------------------

value = "25"

int_value = int(value)
float_value = float(value)
bool_value = bool(int_value)

print("\nType Casting:")
print("Integer:", int_value)
print("Float:", float_value)
print("Boolean:", bool_value)


# --------------------------------------------------
# 10. Demonstrate type casting with list elements
# --------------------------------------------------

str_list = ["1", "2", "3", "4"]

int_list = [int(item) for item in str_list]

print("\nOriginal List:", str_list)
print("Converted to Integer List:", int_list)
```

```python
# ==================================================
# ==================================================

# --------------------------------------------------
# 11. Check if a number is positive, negative, or zero
# --------------------------------------------------

num = 10   # You can change this value to test

if num > 0:
    print("The number is Positive")
elif num < 0:
    print("The number is Negative")
else:
    print("The number is Zero")


# --------------------------------------------------
# 12. Print numbers from 1 to 10 using for loop
# --------------------------------------------------

print("\nNumbers from 1 to 10:")
for i in range(1, 11):
    print(i)


# --------------------------------------------------
# 13. Sum of all even numbers between 1 and 50
# --------------------------------------------------

even_sum = 0

for i in range(1, 51):
    if i % 2 == 0:
        even_sum += i

print("\nSum of even numbers between 1 and 50:", even_sum)


# --------------------------------------------------
# 14. Reverse a string using while loop
# --------------------------------------------------

string = "Python"
reversed_string = ""
index = len(string) - 1

while index >= 0:
    reversed_string += string[index]
    index -= 1

print("\nOriginal String:", string)
print("Reversed String:", reversed_string)


# --------------------------------------------------
# 15. Calculate factorial using while loop
# --------------------------------------------------

number = 5   # Change value to test
factorial = 1
i = 1

while i <= number:
    factorial *= i
    i += 1

print("\nFactorial of", number, "is:", factorial)
```
